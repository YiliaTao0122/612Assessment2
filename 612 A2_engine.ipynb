```python
# Function to find outliers using Z-score for a specific column
def find_outliers_zscore(df, column, threshold=3):
    mean = np.mean(df[column])
    std = np.std(df[column])
    z_scores = (df[column] - mean) / std
    outliers = np.abs(z_scores) > threshold
    return outliers

# Find outliers in 'Units Sold' column using Z-score
outliers = find_outliers_zscore(df, 'Units Sold')

# Print DataFrame with outliers marked
print("DataFrame with outliers in 'Units Sold' column:")
df[outliers]
```

```python
def backup_files(source_dir, backup_path):
    try:
        os.makedirs(backup_path, exist_ok=True)
        print(f"Backup directory created or already exists: {backup_path}")

        all_items = os.listdir(source_dir)
        print(f"Found {len(all_items)} items in source directory: {source_dir}")

        for item in all_items:
            source_item = os.path.join(source_dir, item)
            target_item = os.path.join(backup_path, item)

            try:
                if os.path.isfile(source_item):
                    shutil.copy(source_item, target_item)
                    print(f"Copied {item} to {backup_path}")
                elif os.path.isdir(source_item):
                    shutil.copytree(source_item, target_item)
                    print(f"Copied directory {item} to {backup_path}")
                else:
                    print(f"Skipping {source_item} as it is not a file or directory.")
            except PermissionError:
                print(f"Permission denied: {source_item}")
            except FileNotFoundError:
                print(f"File not found: {source_item}")
            except Exception as e:
                print(f"Error copying {source_item}: {e}")

        print("Backup process completed.")

    except Exception as e:
        print(f"Error creating backup directory or listing items: {e}")

def schedule_backup(source_dir, target_dir_root, target_hour, target_minute, target_second):
    while True:
        now = datetime.now()
        if (now.hour == target_hour and
            now.minute == target_minute and
            now.second == target_second):

            unique_suffix = now.strftime("%Y%m%d%H%M%S")
            backup_path = os.path.join(target_dir_root, f'BACKUP_{unique_suffix}')

            try:
                backup_files(source_dir, backup_path)
                print("Backup completed and task finished.")
                break
            except Exception as e:
                print(f"Backup failed: {e}")
                break

        time.sleep(1)

# Pass the parameters
source_dir = '/content/drive/MyDrive/612 course/612 assessment 2/612 A2 backup'
target_dir_root = '/content/drive/MyDrive/612 course/612 assessment 2/612 A2 backup'
target_hour = 9
target_minute = 9
target_second = 59

schedule_backup(source_dir, target_dir_root, target_hour, target_minute, target_second)
```
